**Advanced Statistical Analysis**

```python

try:
    customers = pd.read_csv("customers.csv")
except:
    customers = None
```

```python

loans.columns = loans.columns.str.strip()
defaults.columns = defaults.columns.str.strip()
if customers is not None:
    customers.columns = customers.columns.str.strip()
```

```python

for col in ["Loan_Amount", "Interest_Rate", "Overdue_Amount", "EMI_Amount"]:
    if col in loans.columns:
        loans[col] = pd.to_numeric(loans[col], errors="coerce")

if customers is not None and "Credit_Score" in customers.columns:
    customers["Credit_Score"] = pd.to_numeric(customers["Credit_Score"], errors="coerce")
```

```python

default_flag = defaults[["Loan_ID"]].drop_duplicates().assign(Default_Flag=1)
default_amount = defaults.groupby("Loan_ID", as_index=False)["Default_Amount"].sum()


loans = loans.merge(default_flag, how="left", on="Loan_ID")
loans["Default_Flag"] = loans["Default_Flag"].fillna(0)

loans = loans.merge(default_amount, how="left", on="Loan_ID")
loans["Default_Amount"] = loans["Default_Amount"].fillna(0)

```

```python

# Calculate Recovery Rate from defaults dataset
# ----------------------------------------
recovery_df = defaults.groupby("Loan_ID").agg({
    "Recovery_Amount": "sum",
    "Default_Amount": "sum"
}).reset_index()

recovery_df["Recovery_Rate"] = recovery_df["Recovery_Amount"] / recovery_df["Default_Amount"].replace(0, np.nan)

loans = loans.merge(recovery_df[["Loan_ID", "Recovery_Rate"]], on="Loan_ID", how="left")
```

```python

# Merge Credit Score (if customers available)
# ----------------------------------------
if customers is not None and "Credit_Score" in customers.columns:
    loans = loans.merge(customers[["Customer_ID", "Credit_Score"]], on="Customer_ID", how="left")
```

```python

# STEP 1  Correlation Analysis for Default Risks
# ============================================================

risk_vars = ["Credit_Score", "Loan_Amount", "Interest_Rate", "Overdue_Amount", "Default_Flag"]

corr_step1 = loans[risk_vars].corr()

print("\n==============================")
print("STEP 1: Default Risk Correlation Matrix")
print("==============================")
print(corr_step1)
```

```python

# Heatmap for Step 1
plt.figure(figsize=(8, 6))
sns.heatmap(corr_step1, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix  Default Risk Variables")
plt.show()
```

```python

# STEP 2  Pairwise Correlation Heatmap
# ============================================================

pair_vars = ["EMI_Amount", "Recovery_Rate", "Default_Amount"]
corr_step2 = loans[pair_vars].corr()

print("\n==============================")
print("STEP 2: EMI / Recovery / Default Correlation Matrix")
print("==============================")
print(corr_step2)
```

```python

# Heatmap for Step 2
plt.figure(figsize=(8, 6))
sns.heatmap(corr_step2, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Heatmap  EMI, Recovery Rate, Default Amount")
plt.show()
```
