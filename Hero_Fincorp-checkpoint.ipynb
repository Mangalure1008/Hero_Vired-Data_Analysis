**7. Transaction and Recovery Analysis**

```python
# ============================================================
# EXECUTE-ALL-IN-ONE ANALYSIS (Inline Charts + Printed Tables)
# Data required in the same folder:
#   - transactions.csv
#   - defaults.csv
#
# What it does:
#   1) Penalty payments & overdue trends (monthly)
#   2) Recovery rates by Default_Reason and by Reason  Legal_Action
#   -> Prints key tables and shows charts inline (plt.show)
# ============================================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# ---------- Styling for charts ----------
sns.set(style="whitegrid", font_scale=1.0)

def load_data():
    """Load and prepare the transactions and defaults datasets."""
    tx = pd.read_csv("transactions.csv")
    df = pd.read_csv("defaults.csv")

    # Dates
    tx["Transaction_Date"] = pd.to_datetime(tx["Transaction_Date"], errors="coerce")
    df["Default_Date"] = pd.to_datetime(df["Default_Date"], errors="coerce")

    # Numerics (safe coercion)
    for c in ["Amount", "Overdue_Fee", "Remaining_Balance"]:
        if c in tx.columns:
            tx[c] = pd.to_numeric(tx[c], errors="coerce")

    for c in ["Default_Amount", "Recovery_Amount"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    return tx, df

# ============================================================
# 1) Penalty payments & overdue trends
# ============================================================
def penalty_overdue_analysis(transactions: pd.DataFrame):
    """Compute monthly penalty trends and overdue fee metrics."""
    # Normalize payment type
    transactions["Payment_Type"] = transactions["Payment_Type"].astype(str).str.lower()

    penalties = transactions[transactions["Payment_Type"] == "penalty"].copy()
    penalties["YearMonth"] = penalties["Transaction_Date"].dt.to_period("M").astype(str)

    all_txn = transactions.copy()
    all_txn["YearMonth"] = all_txn["Transaction_Date"].dt.to_period("M").astype(str)

    # Monthly penalty rollups
    g_pen = penalties.groupby("YearMonth")
    penalty_monthly = pd.DataFrame({
        "Penalty_Transactions": g_pen["Transaction_ID"].size(),
        "Penalty_Amount": g_pen["Amount"].sum(),
        "Avg_Penalty_Amount": g_pen["Amount"].mean()
    }).reset_index()

    # Monthly overdue metrics (all transactions)
    g_all = all_txn.groupby("YearMonth")
    overdue_monthly = pd.DataFrame({
        "Transactions": g_all["Transaction_ID"].size(),
        "Overdue_Fee_Avg": g_all["Overdue_Fee"].mean(),
        "Overdue_Fee_Median": g_all["Overdue_Fee"].median(),
    })
    overdue_monthly["Overdue_Positive_Share"] = g_all["Overdue_Fee"].apply(lambda x: (x.fillna(0) > 0).mean())
    overdue_monthly = overdue_monthly.reset_index()

    # Overdue metrics for penalty transactions only
    g_p = penalties.groupby("YearMonth")
    overdue_penalty_monthly = pd.DataFrame({
        "Penalty_Transactions": g_p["Transaction_ID"].size(),
        "Penalty_Overdue_Avg": g_p["Overdue_Fee"].mean(),
        "Penalty_Overdue_Median": g_p["Overdue_Fee"].median(),
    })
    overdue_penalty_monthly["Penalty_Overdue_Positive_Share"] = g_p["Overdue_Fee"].apply(lambda x: (x.fillna(0) > 0).mean())
    overdue_penalty_monthly = overdue_penalty_monthly.reset_index()

    return penalty_monthly, overdue_monthly, overdue_penalty_monthly

def plot_penalty_overdue(penalty_monthly, overdue_monthly, overdue_penalty_monthly):
    """Show charts inline for penalty trend and overdue trends."""
    # 1) Monthly penalty amount trend
    plt.figure(figsize=(10, 4))
    plt.plot(penalty_monthly["YearMonth"], penalty_monthly["Penalty_Amount"],
             color="#C44E52", marker="o")
    plt.xticks(rotation=45, ha="right")
    plt.title("Monthly Penalty Amount Trend")
    plt.xlabel("Year-Month"); plt.ylabel("Total Penalty Amount")
    plt.tight_layout()
    plt.show()

    # 2) Monthly average overdue fee (all transactions)
    plt.figure(figsize=(10, 4))
    plt.plot(overdue_monthly["YearMonth"], overdue_monthly["Overdue_Fee_Avg"],
             color="#4C72B0", marker="o")
    plt.xticks(rotation=45, ha="right")
    plt.title("Monthly Average Overdue Fee (All Transactions)")
    plt.xlabel("Year-Month"); plt.ylabel("Average Overdue Fee")
    plt.tight_layout()
    plt.show()

    # 3) Share of penalty transactions with overdue fee > 0
    plt.figure(figsize=(10, 4))
    plt.plot(overdue_penalty_monthly["YearMonth"],
             overdue_penalty_monthly["Penalty_Overdue_Positive_Share"],
             color="#55A868", marker="o")
    plt.xticks(rotation=45, ha="right")
    plt.title("Share of Penalty Transactions with Overdue Fee > 0")
    plt.xlabel("Year-Month"); plt.ylabel("Share")
    plt.tight_layout()
    plt.show()

def print_penalty_overdue_headlines(penalty_monthly, overdue_monthly):
    """Print a few headline KPIs to the console."""
    top_penalty_months = penalty_monthly.sort_values("Penalty_Amount", ascending=False).head(5)
    highest_overdue_month = overdue_monthly.sort_values("Overdue_Fee_Avg", ascending=False).head(1)

    print("\n=== Top 5 Months by Total Penalty Amount ===")
    print(top_penalty_months.to_string(index=False))

    print("\n=== Month with Highest Average Overdue Fee (All Txns) ===")
    print(highest_overdue_month.to_string(index=False))

# ============================================================
# 2) Recovery rates by Default_Reason & Legal_Action
# ============================================================
def recovery_analysis(defaults: pd.DataFrame):
    """Compute recovery rates by Default_Reason and by Reason  Legal_Action."""
    # Recovery rate per record (safe division)
    defaults["Recovery_Rate"] = defaults["Recovery_Amount"] / defaults["Default_Amount"].replace(0, np.nan)

    # Reason-level aggregation
    rec_by_reason = (defaults.groupby("Default_Reason")
        .agg(Cases=("Default_ID","count"),
             Total_Default=("Default_Amount","sum"),
             Total_Recovery=("Recovery_Amount","sum"),
             Mean_Recovery_Rate=("Recovery_Rate","mean"),
             Median_Recovery_Rate=("Recovery_Rate","median"))
        .reset_index())
    rec_by_reason["Overall_Recovery_Rate"] = rec_by_reason["Total_Recovery"] / rec_by_reason["Total_Default"].replace(0, np.nan)

    # Reason  Legal_Action aggregation
    rec_by_reason_legal = (defaults.groupby(["Default_Reason","Legal_Action"])
        .agg(Cases=("Default_ID","count"),
             Total_Default=("Default_Amount","sum"),
             Total_Recovery=("Recovery_Amount","sum"),
             Mean_Recovery_Rate=("Recovery_Rate","mean"),
             Median_Recovery_Rate=("Recovery_Rate","median"))
        .reset_index())
    rec_by_reason_legal["Overall_Recovery_Rate"] = rec_by_reason_legal["Total_Recovery"] / rec_by_reason_legal["Total_Default"].replace(0, np.nan)

    return rec_by_reason, rec_by_reason_legal

def plot_recovery(rec_by_reason, rec_by_reason_legal):
    """Show charts inline for recovery rates by reason and by reason  legal."""
    # 1) Recovery rate by default reason (bar)
    plt.figure(figsize=(9, 4))
    ordered = rec_by_reason.sort_values("Overall_Recovery_Rate", ascending=False)
    sns.barplot(data=ordered, x="Default_Reason", y="Overall_Recovery_Rate", color="#8172B3")
    plt.title("Overall Recovery Rate by Default Reason")
    plt.xlabel("Default Reason"); plt.ylabel("Overall Recovery Rate")
    plt.xticks(rotation=30, ha="right")
    plt.tight_layout()
    plt.show()

    # 2) Recovery rate by reason  legal action (clustered bar)
    pivot = rec_by_reason_legal.pivot(index="Default_Reason", columns="Legal_Action", values="Overall_Recovery_Rate")
    pivot = pivot.sort_index()  # keep the order consistent
    ax = pivot.plot(kind="bar", figsize=(10, 5))
    ax.set_title("Overall Recovery Rate by Reason & Legal Action")
    ax.set_xlabel("Default Reason"); ax.set_ylabel("Overall Recovery Rate")
    plt.tight_layout()
    plt.show()

def print_recovery_headlines(rec_by_reason, defaults):
    """Print a few headline KPIs to the console."""
    best_reason = rec_by_reason.sort_values("Overall_Recovery_Rate", ascending=False).head(1)
    legal_lift = (defaults.groupby("Legal_Action")
                  .agg(Total_Default=("Default_Amount","sum"),
                       Total_Recovery=("Recovery_Amount","sum"))
                  .assign(Overall_Recovery_Rate=lambda d: d["Total_Recovery"]/d["Total_Default"].replace(0,np.nan)))

    print("\n=== Best Recovery Reason (Overall) ===")
    print(best_reason.to_string(index=False))

    print("\n=== Overall Recovery Rate by Legal Action ===")
    print(legal_lift.to_string())

# ============================================================
# MAIN: Run everything inline
# ============================================================
if __name__ == "__main__":
    transactions, defaults = load_data()

    # ---- Part 1: Penalty + Overdue ----
    penalty_monthly, overdue_monthly, overdue_penalty_monthly = penalty_overdue_analysis(transactions)
    # Print tables (headlines + quick previews)
    print_penalty_overdue_headlines(penalty_monthly, overdue_monthly)

    print("\n=== Penalty Monthly (preview) ===")
    print(penalty_monthly.head(10).to_string(index=False))

    print("\n=== Overdue Monthly (preview) ===")
    print(overdue_monthly.head(10).to_string(index=False))

    print("\n=== Overdue (Penalties Only) Monthly (preview) ===")
    print(overdue_penalty_monthly.head(10).to_string(index=False))

    # Show charts inline
    plot_penalty_overdue(penalty_monthly, overdue_monthly, overdue_penalty_monthly)

    # ---- Part 2: Recovery ----
    rec_by_reason, rec_by_reason_legal = recovery_analysis(defaults)
    print_recovery_headlines(rec_by_reason, defaults)

    print("\n=== Recovery by Reason (preview) ===")
    print(rec_by_reason.to_string(index=False))

    print("\n=== Recovery by Reason  Legal Action (preview) ===")
    print(rec_by_reason_legal.to_string(index=False))

    # Show charts inline
    plot_recovery(rec_by_reason, rec_by_reason_legal)

    print("\nAll analyses complete. Charts were displayed inline. Tables were printed above.")
```

**9.Loan Application Insights**

```python
# ============================================================
# Loan Application Insights (Inline Charts + Printed Outputs)
# Files expected in the same folder:
#   - applications.csv
#   - loans.csv
#
# Outputs:
#   1) Approval & rejection rates
#   2) Top rejection reasons
#   3) Processing fee comparison (approved vs rejected)
# Charts are shown inline with plt.show()
# ============================================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# ---- Chart style ----
sns.set(style="whitegrid", font_scale=1.0)

# ---- Load datasets ----
applications = pd.read_csv("applications.csv")
loans = pd.read_csv("loans.csv")  # Loaded per request; not required for these 3 insights

# ---- Basic cleaning ----
for df in (applications, loans):
    df.columns = df.columns.str.strip()

# Normalize fields
applications["Approval_Status"] = applications["Approval_Status"].astype(str).str.strip().str.title()
if "Rejection_Reason" in applications.columns:
    applications["Rejection_Reason"] = applications["Rejection_Reason"].astype(str).str.strip()

# Numeric conversion for Processing_Fee
if "Processing_Fee" in applications.columns:
    applications["Processing_Fee"] = pd.to_numeric(applications["Processing_Fee"], errors="coerce")

# ============================================================
# 1) Approval & Rejection rates
# ============================================================
status_counts = applications["Approval_Status"].value_counts(dropna=False)
status_total = int(status_counts.sum())
approved_n = int(status_counts.get("Approved", 0))
rejected_n = int(status_counts.get("Rejected", 0))
other_n = status_total - approved_n - rejected_n

approved_rate = approved_n / status_total if status_total else np.nan
rejected_rate = rejected_n / status_total if status_total else np.nan

print("\n=== Approval & Rejection Summary ===")
print({
    "Total_Applications": status_total,
    "Approved": approved_n,
    "Rejected": rejected_n,
    "Other_Status": other_n,
    "Approval_Rate": approved_rate,
    "Rejection_Rate": rejected_rate
})

# Chart: approved vs rejected
plt.figure(figsize=(6,4))
sns.barplot(x=["Approved","Rejected"], y=[approved_n, rejected_n], palette=["#55A868","#C44E52"])
plt.title("Application Outcomes: Approved vs Rejected")
plt.xlabel("Outcome"); plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ============================================================
# 2) Most common rejection reasons
# ============================================================
rej_df = applications[applications["Approval_Status"] == "Rejected"].copy()
reasons = (rej_df.groupby("Rejection_Reason")
           .size()
           .reset_index(name="Count")
           .sort_values("Count", ascending=False))

# Add share column against total rejected
reasons["Share"] = reasons["Count"] / len(rej_df) if len(rej_df) > 0 else 0.0

print("\n=== Top Rejection Reasons (Top 10) ===")
print(reasons.head(10).to_string(index=False))

# Chart: Top reasons (Top-10)
top_k = min(10, len(reasons))
plt.figure(figsize=(8,4))
sns.barplot(data=reasons.head(top_k), x="Rejection_Reason", y="Count", color="#8172B3")
plt.title("Top Rejection Reasons")
plt.xlabel("Reason"); plt.ylabel("Count")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

# ============================================================
# 3) Processing fee comparison (Approved vs Rejected)
# ============================================================
fee_stats = (applications.dropna(subset=["Processing_Fee"])
             .groupby("Approval_Status")["Processing_Fee"]
             .agg(["count","mean","median","std"])
             .reset_index())

print("\n=== Processing Fee Comparison by Outcome ===")
print(fee_stats.to_string(index=False))

# Chart: fee distribution by status
plt.figure(figsize=(8,4))
sns.boxplot(data=applications.dropna(subset=["Processing_Fee"]),
            x="Approval_Status", y="Processing_Fee")
plt.title("Processing Fee Distribution by Application Outcome")
plt.xlabel("Outcome"); plt.ylabel("Processing Fee")
plt.tight_layout()
plt.show()

print("\nAnalysis complete: charts shown inline; numeric summaries printed above.")
```

**10. Recovery Effectiveness**

```python
# ============================================================
# Recovery Effectiveness + Legal Action Comparison + Branch Ops
# Files expected in the same folder:
#   - defaults.csv
#   - branches.csv
# Charts shown inline (plt.show()) and also saved with simple names
# ============================================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid', font_scale=1.0)

# Load
defaults = pd.read_csv('defaults.csv')
branches = pd.read_csv('branches.csv')

# Clean
for df in (defaults, branches):
    df.columns = df.columns.str.strip()

for c in ['Default_Amount','Recovery_Amount']:
    if c in defaults.columns:
        defaults[c] = pd.to_numeric(defaults[c], errors='coerce')

# Recovery rate per default
defaults['Recovery_Rate'] = defaults['Recovery_Amount'] / defaults['Default_Amount'].replace(0, np.nan)

# -------------------------
# 1) Recovery Effectiveness
# -------------------------
overall_total_default = defaults['Default_Amount'].sum(skipna=True)
overall_total_recovery = defaults['Recovery_Amount'].sum(skipna=True)
overall_recovery_rate = (overall_total_recovery / overall_total_default) if overall_total_default else np.nan

mean_rr   = defaults['Recovery_Rate'].mean(skipna=True)
median_rr = defaults['Recovery_Rate'].median(skipna=True)
count_rr  = defaults['Recovery_Rate'].count()

print("\n=== Overall Recovery Effectiveness ===")
print({
    "records_with_recovery_rate": int(count_rr),
    "overall_recovery_rate_totals": float(overall_recovery_rate) if pd.notna(overall_recovery_rate) else None,
    "mean_recovery_rate_per_record": float(mean_rr) if pd.notna(mean_rr) else None,
    "median_recovery_rate_per_record": float(median_rr) if pd.notna(median_rr) else None
})

# Chart: recovery-rate distribution
plt.figure(figsize=(8,4))
sns.histplot(defaults['Recovery_Rate'].dropna(), bins=40, color='#4C72B0')
plt.title('Distribution of Recovery Rate (Recovery_Amount / Default_Amount)')
plt.xlabel('Recovery Rate'); plt.ylabel('Count')
plt.tight_layout()
plt.savefig('recovery_rate_distribution.png', dpi=150)
plt.show()
plt.close()

# -----------------------------------
# 2) Legal vs Non-Legal Comparisons
# -----------------------------------
legal_cols = [c for c in defaults.columns if c.lower() == 'legal_action']
if legal_cols:
    legal_col = legal_cols[0]
    defaults[legal_col] = defaults[legal_col].astype(str).str.strip().str.title()

    by_legal = (defaults.groupby(legal_col)
                .agg(Cases=('Default_ID','count'),
                     Total_Default=('Default_Amount','sum'),
                     Total_Recovery=('Recovery_Amount','sum'),
                     Mean_Recovery_Rate=('Recovery_Rate','mean'),
                     Median_Recovery_Rate=('Recovery_Rate','median'))
                .reset_index())
    by_legal['Overall_Recovery_Rate'] = by_legal['Total_Recovery'] / by_legal['Total_Default'].replace(0, np.nan)

    print("\n=== Recovery by Legal Action ===")
    print(by_legal.to_string(index=False))

    # Chart: overall recovery rate by legal action
    plt.figure(figsize=(7,4))
    sns.barplot(data=by_legal, x=legal_col, y='Overall_Recovery_Rate', color='#55A868')
    plt.title('Overall Recovery Rate by Legal Action')
    plt.xlabel('Legal Action'); plt.ylabel('Overall Recovery Rate')
    plt.tight_layout()
    plt.savefig('recovery_by_legal_action.png', dpi=150)
    plt.show()
    plt.close()
else:
    print("\n=== Recovery by Legal Action ===")
    print("Legal_Action column not found in defaults.csv; unable to compare legal vs non-legal.")

# -----------------------------------
# 3) Branch-wise Recovery Performance
# -----------------------------------
# We need a mapping Loan_ID -> Branch_ID (or Region) to compute true branch recovery.
# Since it's missing, we print branch operational metrics to provide context.
branch_ops = branches[['Branch_ID','Region','Total_Active_Loans','Delinquent_Loans','Loan_Disbursement_Amount']].copy()
print("\n=== Branch Operational Metrics (available) ===")
print(branch_ops.head(10).to_string(index=False))

# Chart: Top 10 branches by Delinquent_Loans (operational)
plt.figure(figsize=(10,4))
plot_df = branch_ops.sort_values('Delinquent_Loans', ascending=False).head(10)
sns.barplot(data=plot_df, x='Branch_ID', y='Delinquent_Loans', color='#C44E52')
plt.title('Top 10 Branches by Delinquent Loans (Operational)')
plt.xlabel('Branch'); plt.ylabel('Delinquent Loans')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top10_branches_delinquents.png', dpi=150)
plt.show()
plt.close()
```

**14. Default Trends**

```python
# ============================================================
# Default Trends (Inline Charts + Printed Outputs)
# Requires: defaults.csv
# Optional enrichments: loans.csv (for Loan_Purpose), customers.csv (for Income categories)
# ============================================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid', font_scale=1.0)

# ------------------------- Load -------------------------
defaults = pd.read_csv('defaults.csv')
try:
    loans = pd.read_csv('loans.csv')
except:
    loans = None
try:
    customers = pd.read_csv('customers.csv')
except:
    customers = None

# ------------------------- Clean & Types -------------------------
for df in [d for d in [defaults, loans, customers] if d is not None]:
    df.columns = df.columns.str.strip()

defaults['Default_Date'] = pd.to_datetime(defaults['Default_Date'], errors='coerce')
for c in ['Default_Amount','Recovery_Amount']:
    if c in defaults.columns:
        defaults[c] = pd.to_numeric(defaults[c], errors='coerce')

# ============================================================
# 1) Defaults over time (counts & amounts)
# ============================================================
defaults['Default_Month'] = defaults['Default_Date'].dt.to_period('M').astype(str)

monthly = (defaults.dropna(subset=['Default_Month'])
           .groupby('Default_Month')
           .agg(Defaults=('Default_ID','nunique'),
                Total_Default_Amount=('Default_Amount','sum'))
           .reset_index()
           .sort_values('Default_Month'))

monthly['Defaults_MA3'] = monthly['Defaults'].rolling(3, min_periods=1).mean()

# Chart: Number of defaults over time
plt.figure(figsize=(10,4))
plt.plot(monthly['Default_Month'], monthly['Defaults'], color='#4C72B0', label='Defaults')
plt.plot(monthly['Default_Month'], monthly['Defaults_MA3'], color='#55A868', linestyle='--', label='3M MA')
plt.xticks(rotation=45, ha='right')
plt.title('Number of Defaults Over Time')
plt.xlabel('Month'); plt.ylabel('Defaults'); plt.legend()
plt.tight_layout()
plt.show()

# Chart: Total default amount over time
plt.figure(figsize=(10,4))
sns.barplot(data=monthly, x='Default_Month', y='Total_Default_Amount', color='#C44E52')
plt.xticks(rotation=45, ha='right')
plt.title('Total Default Amount Over Time')
plt.xlabel('Month'); plt.ylabel('Total Default Amount')
plt.tight_layout()
plt.show()

# ============================================================
# 2) Average default amount by loan purpose
# ============================================================
purpose_col = None
if loans is not None:
    for c in ['Loan_Purpose','Loan_Type','Product_Type','Loan_Product','Category','Loan_Segment']:
        if c in loans.columns:
            purpose_col = c
            break

avg_by_purpose = None
if loans is not None and purpose_col is not None:
    join_df = defaults.merge(loans[['Loan_ID', purpose_col]], how='left', on='Loan_ID')
    avg_by_purpose = (join_df.dropna(subset=[purpose_col])
                      .groupby(purpose_col)
                      .agg(Average_Default_Amount=('Default_Amount','mean'),
                           Defaults=('Default_ID','nunique'))
                      .reset_index()
                      .sort_values('Average_Default_Amount', ascending=False))

    plt.figure(figsize=(10,4))
    sns.barplot(data=avg_by_purpose, x=purpose_col, y='Average_Default_Amount', color='#8172B3')
    plt.title('Average Default Amount by Loan Purpose')
    plt.xlabel('Loan Purpose'); plt.ylabel('Avg Default Amount')
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.show()

# ============================================================
# 3) Default rates across income categories
# ============================================================
income_cat_col = None
numeric_income_col = None
if customers is not None:
    for c in ['Income_Category','Income_Band','IncomeBracket','Income_Bucket']:
        if c in customers.columns:
            income_cat_col = c
            break
    if income_cat_col is None:
        for c in ['Annual_Income','Income','Monthly_Income']:
            if c in customers.columns:
                numeric_income_col = c
                customers[c] = pd.to_numeric(customers[c], errors='coerce')
                break
    if income_cat_col is None and numeric_income_col is not None:
        # quantile buckets
        qlabels = ['Low','Lower-Mid','Upper-Mid','High']
        customers['Income_Category_Derived'] = pd.qcut(
            customers[numeric_income_col].rank(method='first'), q=4, labels=qlabels
        )
        income_cat_col = 'Income_Category_Derived'

rate_by_income = None
if (loans is not None) and (customers is not None) and (income_cat_col is not None):
    default_flag = defaults[['Loan_ID']].drop_duplicates().assign(Default_Flag=1)
    loans2 = loans[['Loan_ID','Customer_ID']].drop_duplicates()
    loans2 = loans2.merge(default_flag, how='left', on='Loan_ID')
    loans2['Default_Flag'] = loans2['Default_Flag'].fillna(0).astype(int)

    join2 = loans2.merge(customers[['Customer_ID', income_cat_col]], how='left', on='Customer_ID')

    rate_by_income = (join2.dropna(subset=[income_cat_col])
                      .groupby(income_cat_col)
                      .agg(Loans=('Loan_ID','nunique'),
                           Defaults=('Default_Flag','sum'))
                      .assign(Default_Rate=lambda d: d['Defaults']/d['Loans'])
                      .reset_index()
                      .sort_values('Default_Rate', ascending=False))

    plt.figure(figsize=(8,4))
    sns.barplot(data=rate_by_income, x=income_cat_col, y='Default_Rate', color='#55A868')
    plt.title('Default Rate by Income Category')
    plt.xlabel('Income Category'); plt.ylabel('Default Rate')
    plt.tight_layout()
    plt.show()

# ------------------------- Print quick tables -------------------------
print("\n=== Monthly Defaults (head) ===")
print(monthly.head().to_string(index=False))

if avg_by_purpose is not None:
    print("\n=== Average Default Amount by Loan Purpose ===")
    print(avg_by_purpose.to_string(index=False))
else:
    print("\n(No loan purpose column found in loans.csv; unable to compute purpose-wise averages.)")

if rate_by_income is not None:
    print("\n=== Default Rate by Income Category ===")
    print(rate_by_income.to_string(index=False))
else:
    print("\n(Income categories not found; provide customers.csv with Income or Income_Category to compute rates.)")
```
